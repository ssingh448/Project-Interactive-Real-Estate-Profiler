```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/bishalregmi/Downloads/RDC_Inventory_Hotness_Metrics_Zip_History.csv", dtype='object')[:-1]
df.head(5)
```

```python
df['postal_code'] = df['postal_code'].str.zfill(5)
```

```python
df['month_date_yyyymm'] = pd.to_datetime(df['month_date_yyyymm'], format='%Y%m')
df.sort_values(['postal_code', 'month_date_yyyymm'], inplace=True)
most_recent_idx = df.groupby('postal_code')['month_date_yyyymm'].idxmax()
most_recent_rows = df.loc[most_recent_idx]
most_recent_rows['postal_code'] = most_recent_rows['postal_code'].str.zfill(5)
```

```python
nj_df = most_recent_rows[most_recent_rows['postal_code'].str.startswith(('07', '08'))]
ny_df = most_recent_rows[most_recent_rows['postal_code'].str.startswith(('10', '11', '12', '13', '14'))]
ct_df = most_recent_rows[most_recent_rows['postal_code'].str.startswith(('06'))]
```

```python
all_df = pd.concat([nj_df, ny_df, ct_df], axis=0)
all_df
```
